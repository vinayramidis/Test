## Step 2: Data Cleaning and Preparation

```python
#Splitting company name from CarName column
CompanyName = cp['CarName'].apply(lambda x : x.split(' ')[0])
cp.insert(3,"CompanyName",CompanyName)
cp.drop(['CarName'],axis=1,inplace=True)
cp.head()
```

Fixing invalid values

There seems to be some spelling error in the CompanyName column.

    maxda = mazda
    Nissan = nissan
    porsche = porcshce
    toyota = toyouta
    vokswagen = volkswagen = vw

```python
cp.CompanyName.unique()
```

```python
cp.CompanyName = cp.CompanyName.str.lower()

def replace_name(a,b):
    cp.CompanyName.replace(a,b,inplace=True)

replace_name('maxda','mazda')
replace_name('porcshce','porsche')
replace_name('toyouta','toyota')
replace_name('vokswagen','volkswagen')
replace_name('vw','volkswagen')

cp.CompanyName.unique()
```

## Step 3: Visualizing the data

```python
plt.figure(figsize=(20,8))

plt.subplot(1,2,1)
plt.title('Car Price Distribution Plot')
sns.distplot(cp.price)

plt.subplot(1,2,2)
plt.title('Car Price Spread')
sns.boxplot(y=cp.price)

plt.show()
```

```python
print(cp.price.describe(percentiles = [0.25,0.50,0.75,0.85,0.90,1]))
```

Inference :
The plot seemed to be right-skewed, meaning that the 75% of prices in the dataset are low(Below 16,503).

There is a significant difference between the mean and the median of the price distribution.

The data points are far spread out from the mean, which indicates a high variance in the car prices.(85% of the prices are below 18,500, whereas the remaining 15% are between 18,500 and 45,400.)
